```python
def partition(arr, low, high):
    i = (low - 1)
    pivot = arr[high]

    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i+1], arr[high] = arr[high], arr[i+1]
    return (i+1)


def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi-1)
        quick_sort(arr, pi+1, high)
```

```python
def strassen_matrix_multiply(A, B):
    # Base case: if the matrices are 1x1, just multiply them
    if len(A) == 1:
        return [[A[0][0] * B[0][0]]]
    
    # Split the matrices into quadrants
    n = len(A)
    A11 = [row[:n//2] for row in A[:n//2]]
    A12 = [row[n//2:] for row in A[:n//2]]
    A21 = [row[:n//2] for row in A[n//2:]]
    A22 = [row[n//2:] for row in A[n//2:]]
    B11 = [row[:n//2] for row in B[:n//2]]
    B12 = [row[n//2:] for row in B[:n//2]]
    B21 = [row[:n//2] for row in B[n//2:]]
    B22 = [row[n//2:] for row in B[n//2:]]
    
    # Recursively compute the seven products needed for the final result
    M1 = strassen_matrix_multiply(add_matrices(A11, A22), add_matrices(B11, B22))
    M2 = strassen_matrix_multiply(add_matrices(A21, A22), B11)
    M3 = strassen_matrix_multiply(A11, subtract_matrices(B12, B22))
    M4 = strassen_matrix_multiply(A22, subtract_matrices(B21, B11))
    M5 = strassen_matrix_multiply(add_matrices(A11, A12), B22)
    M6 = strassen_matrix_multiply(subtract_matrices(A21, A11), add_matrices(B11, B12))
    M7 = strassen_matrix_multiply(subtract_matrices(A12, A22), add_matrices(B21, B22))
    
    # Combine the results to get the final product
    C11 = add_matrices(subtract_matrices(add_matrices(M1, M4), M5), M7)
    C12 = add_matrices(M3, M5)
    C21 = add_matrices(M2, M4)
    C22 = add_matrices(subtract_matrices(add_matrices(M1, M3), M2), M6)
    C = [[0] * n for _ in range(n)]
    for i in range(n//2):
        C[i][:n//2] = C11[i]
        C[i][n//2:] = C12[i]
        C[i+n//2][:n//2] = C21[i]
        C[i+n//2][n//2:] = C22[i]
    return C

def add_matrices(A, B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def subtract_matrices(A, B):
    return [[A[i][j] - B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

```

```python
def activity_selection(start, end):
    n = len(start)
    activities = []
    for i in range(n):
        activities.append((start[i], end[i]))
    activities.sort(key=lambda x: x[1])  # sort activities by end time

    selected_activities = [activities[0]]
    last_end_time = activities[0][1]

    for activity in activities[1:]:
        start_time, end_time = activity
        if start_time >= last_end_time:
            selected_activities.append(activity)
            last_end_time = end_time

    return selected_activities


# Example Usage
start = [1, 3, 0, 5, 8, 5]
end = [2, 4, 6, 7, 9, 9]
selected_activities = activity_selection(start, end)
print("Selected Activities:", selected_activities)
```

```python
from heapq import heappush, heappop
from math import inf

def dijkstra(graph, start):
    distances = {start: 0}
    previous = {start: None}
    heap = [(0, start)]

    while heap:
        (distance, current) = heappop(heap)

        if distance > distances[current]:
            continue

        for neighbor, weight in graph[current]:
            tentative_distance = distance + weight
            if tentative_distance < distances.get(neighbor, inf):
                distances[neighbor] = tentative_distance
                previous[neighbor] = current
                heappush(heap, (tentative_distance, neighbor))

    return distances, previous
```

```python
def longest_common_subsequence(X, Y):
    m = len(X)
    n = len(Y)
    L = [[0] * (n+1) for i in range(m+1)]
    for i in range(1, m+1):
        for j in range(1, n+1):
            if X[i-1] == Y[j-1]:
                L[i][j] = L[i-1][j-1] + 1
            else:
                L[i][j] = max(L[i-1][j], L[i][j-1])
            return L[m][n]
```

```python
import heapq

def dijkstra(graph, start):
    distances = {node: float("inf") for node in graph}
    distances[start] = 0

    pq = [(0, start)]

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return distances
```

```python
def bellman_ford(graph, start):
    distances = {node: float("inf") for node in graph}
    distances[start] = 0

    for _ in range(len(graph) - 1):
        for node in graph:
            for neighbor, weight in graph[node].items():
                if distances[node] + weight < distances[neighbor]:
                    distances[neighbor] = distances[node] + weight

    # check for negative cycles
    for node in graph:
        for neighbor, weight in graph[node].items():
            if distances[node] + weight < distances[neighbor]:
                raise ValueError("Negative cycle detected")

    return distances
```

```python
def kruskal(graph):
    mst = []
    parents = {}
    ranks = {}

    def find(node):
        if parents[node] != node:
            parents[node] = find(parents[node])
        return parents[node]

    def union(node1, node2):
        root1 = find(node1)
        root2 = find(node2)

        if root1 != root2:
            if ranks[root1] > ranks[root2]:
                parents[root2] = root1
            else:
                parents[root1] = root2
                if ranks[root1] == ranks[root2]:
                    ranks[root2] += 1

    edges = []
    for node in graph:
        parents[node] = node
        ranks[node] = 0
        for neighbor, weight in graph[node].items():
            edges.append((weight, node, neighbor))

    edges.sort()

    for edge in edges:
        weight, node1, node2 = edge
        if find(node1) != find(node2):
            union(node1, node2)
            mst.append(edge)

    return mst
```

```python
def bubble_sort(arr):
    n = len(arr)
    
    # Traverse through all array elements
    for i in range(n):
        
        # Last i elements are already sorted
        for j in range(0, n-i-1):
            
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    
    return arr
```

```python
def binary_search(arr, target):
    # Set initial search range
    low = 0
    high = len(arr) - 1
    
    # While the search range is valid
    while low <= high:
        mid = (low + high) // 2
        
        # Check if target is present at mid
        if arr[mid] == target:
            return mid
        
        # If target is greater, ignore left half
        elif arr[mid] < target:
            low = mid + 1
        
        # If target is smaller, ignore right half
        else:
            high = mid - 1
    
    # If we reach here, the element was not present
    return -1

```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        left_arr = []
        right_arr = []
        equal_arr = []
        for elem in arr:
            if elem < pivot:
                left_arr.append(elem)
            elif elem > pivot:
                right_arr.append(elem)
            else:
                equal_arr.append(elem)
        return quick_sort(left_arr) + equal_arr + quick_sort(right_arr)
```

```python
def heapify(arr, n, i):
    largest = i
    left = 2*i + 1
    right = 2*i + 2

    if left < n and arr[largest] < arr[left]:
        largest = left

    if right < n and arr[largest] < arr[right]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort(arr):
    n = len(arr)

    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

    return arr
```

```python
import heapq

def prim(graph):
    # Initialize
    visited = set()
    edges = []
    start = next(iter(graph))
    visited.add(start)
    for e in graph[start]:
        heapq.heappush(edges, e)

    # Build the MST
    mst = []
    while edges:
        # Get the minimum edge
        weight, v1, v2 = heapq.heappop(edges)
        if v2 not in visited:
            visited.add(v2)
            mst.append((v1, v2, weight))
            # Add new edges to the heap
            for e in graph[v2]:
                if e[2] not in visited:
                    heapq.heappush(edges, e)

    return mst
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    else:
        fib = [0] * (n+1)
        fib[0] = 0
        fib[1] = 1
        for i in range(2, n+1):
            fib[i] = fib[i-1] + fib[i-2]
        return fib[n]
```

```python
def knapsack(W, wt, val, n):
    K = [[0 for x in range(W + 1)] for x in range(n + 1)]
    
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                K[i][w] = 0
            elif wt[i-1] <= w:
                K[i][w] = max(val[i-1] + K[i-1][w-wt[i-1]],  K[i-1][w])
            else:
                K[i][w] = K[i-1][w]
    return K[n][W]

W = 50
val = [60, 100, 120]
wt = [10, 20, 30]
n = len(val)
print(knapsack(W, wt, val, n)) # Output: 220
```

```python
def fractional_knapsack(values, weights, capacity):
    n = len(values)
    items = [(values[i], weights[i]) for i in range(n)]
    items.sort(reverse=True, key=lambda x: x[0] / x[1])

    total_value = 0
    for item in items:
        if capacity == 0:
            break
        value_per_weight = item[0] / item[1]
        weight = min(item[1], capacity)
        total_value += value_per_weight * weight
        capacity -= weight

    return total_value
```

```python
def is_valid(board, row, col):
    for i in range(row):
        if board[i] == col or abs(board[i]-col) == row-i:
            return False
    return True

def solve_n_queens(n):
    board = [-1]*n
    solutions = []

    def backtrack(row):
        nonlocal solutions
        if row == n:
            solutions.append(board[:])
            return
        for col in range(n):
            if is_valid(board, row, col):
                board[row] = col
                backtrack(row+1)
                board[row] = -1
    
    backtrack(0)
    return solutions
```

```python
def subset_sum(nums, target):
    def backtrack(start, cur_sum, path):
        nonlocal result
        if cur_sum == target:
            result = path[:]
            return
        if cur_sum > target or start == len(nums):
            return
        for i in range(start, len(nums)):
            path.append(nums[i])
            backtrack(i+1, cur_sum+nums[i], path)
            path.pop()
    
    result = []
    nums.sort()
    backtrack(0, 0, [])
    return result
```

```python
def hamiltonian_cycle(graph):
    n = len(graph)
    path = [-1]*n

    def is_valid(vertex, pos):
        if graph[path[pos-1]][vertex] == 0:
            return False
        if vertex in path[:pos]:
            return False
        return True

    def backtrack(pos):
        if pos == n:
            if graph[path[pos-1]][path[0]] == 1:
                return True
            return False
        for vertex in range(1, n):
            if is_valid(vertex, pos):
                path[pos] = vertex
                if backtrack(pos+1):
                    return True
                path[pos] = -1
        return False

    path[0] = 0
    if backtrack(1):
        return path
    return []
```

```python
import heapq
from collections import defaultdict

def huffman_encoding(data):
    if not data:
        return "", None

    freq = defaultdict(int)
    for c in data:
        freq[c] += 1

    # Build a min heap of the character frequencies
    heap = [[freq[c], c] for c in freq]
    heapq.heapify(heap)

    # Combine two smallest frequencies until there is only one left
    while len(heap) > 1:
        freq1, c1 = heapq.heappop(heap)
        freq2, c2 = heapq.heappop(heap)
        merged_freq = freq1 + freq2
        heapq.heappush(heap, [merged_freq, [c1, c2]])

    # Traverse the Huffman tree and assign codes to each character
    huffman_code = {}
    def assign_codes(node, code):
        if isinstance(node, list):
            assign_codes(node[0], code + "0")
            assign_codes(node[1], code + "1")
        else:
            huffman_code[node] = code

    root = heapq.heappop(heap)[1]
    assign_codes(root, "")

    # Encode the original data using the Huffman codes
    encoded_data = "".join(huffman_code[c] for c in data)

    return encoded_data, root


def huffman_decoding(data, tree):
    if not data:
        return ""

    # Traverse the Huffman tree and decode the encoded data
    decoded_data = []
    node = tree
    for bit in data:
        if bit == "0":
            node = node[0]
        else:
            node = node[1]

        if isinstance(node, str):
            decoded_data.append(node)
            node = tree

    return "".join(decoded_data)


# Example Usage
data = "The quick brown fox jumps over the lazy dog"
encoded_data, tree = huffman_encoding(data)
print("Encoded Data:", encoded_data)
decoded_data = huffman_decoding(encoded_data, tree)
print("Decoded Data:", decoded_data)
```
